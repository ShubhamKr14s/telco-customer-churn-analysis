# Task 3: Exploratory Data Analysis (EDA)
Objective: To analyze trends, distributions, and churn behavior using statistical methods and visualizations.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Telco_Customer_Churn_Dataset  (3).csv")

# Basic cleaning for analysis
df.columns = df.columns.str.lower().str.replace(" ", "_")
df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
df['totalcharges'] = df['totalcharges'].fillna(df['totalcharges'].median())
```

```python
df.describe()
```

```python
sns.countplot(x='churn', data=df)
plt.title("Churn Distribution")
plt.savefig("churn_distribution.png")
plt.show()
```

```python
df['monthlycharges'].hist(bins=30)
plt.title("Monthly Charges Distribution")
plt.savefig("monthly_charges_histogram.png")
plt.show()
```

```python
sns.boxplot(x='churn', y='monthlycharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.savefig("charges_vs_churn_boxplot.png")
plt.show()
```

```python
df['tenure'].hist(bins=30)
plt.title("Tenure Distribution")
plt.savefig("tenure_distribution.png")
plt.show()
```

## Key Insights

- Majority customers are non-churned.
- Customers with higher monthly charges tend to churn more.
- Short tenure customers show higher churn tendency.
- Customer base includes both short-term and long-term subscribers.
