# Task 4: Customer Segmentation Visualization
Objective: To segment customers based on tenure and analyze charge patterns across segments.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Telco_Customer_Churn_Dataset  (3).csv")

# Basic cleaning
df.columns = df.columns.str.lower().str.replace(" ", "_")
df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
df['totalcharges'] = df['totalcharges'].fillna(df['totalcharges'].median())
```

```python
bins = [0, 12, 36, 72]
labels = ['0-12 Months', '13-36 Months', '37+ Months']

df['tenure_group'] = pd.cut(df['tenure'], bins=bins, labels=labels)
df['tenure_group'].value_counts()
```

```python
df['tenure_group'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Customer Distribution by Tenure Group")
plt.ylabel("")
plt.savefig("tenure_pie_chart.png")
plt.show()
```

```python
grouped = df.groupby('tenure_group')['monthlycharges'].mean()

grouped.plot(kind='bar')
plt.title("Average Monthly Charges by Tenure Group")
plt.ylabel("Average Monthly Charges")
plt.savefig("tenure_bar_chart.png")
plt.show()
```

## Segmentation Insights

- A significant portion of customers fall in the early tenure group (0-12 months).
- Long-term customers (37+ months) form a stable customer base.
- Monthly charges vary across tenure groups.
- Understanding tenure segments helps in targeted retention strategies.
